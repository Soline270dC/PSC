```python
from time_series_gan.GAN import GAN
import numpy as np
import pandas as pd
```

```python
model = GAN()
data = np.loadtxt("data/data_gdp.csv", skiprows = 1)
model.set_data(pd.DataFrame(data))
model.set_architecture()
```

```python
model.fit_hyperparameters({"lr_g": [0.5e-5, 5e-4], "lr_d": [0.5e-5, 5e-4], "epochs": [50, 150], "batch_size": [8,100], "latent_dim": [50,150]})
```

```python
model.fit(verbose = True)
```
